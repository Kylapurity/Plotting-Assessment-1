```python
# YOUR CODE HERE
x_min, x_max = 0, 700
x_step = 70
bin_edges = np.arange(x_min, x_max + x_step, x_step)
plt.figure(figsize=(8, 6))
plt.hist(data['co2'], bins=20, color='brown', edgecolor='black')
plt.xticks(np.arange(x_min, x_max + x_step, x_step))
plt.title('Distribution of carbon Dioxide')
plt.xlabel('co2 (mpg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(8, 6))
plt.hist(data['city'], bins=20, color='Green', edgecolor='black')  
plt.title('City Data')
plt.xlabel('City(mpg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# DataFrame
correlation_heatmap = data.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_heatmap, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap Correlation")
plt.show()
```

HEATMAP: Understanding  Representation

The heatmap provides a visual representation of the correlations between different variables in your dataset. Here's how to interpret the key elements:

Color Intensity: The color intensity represents the strength of the correlation:

Dark red: Strong positive correlation Dark blue: Strong negative correlation Lighter colors: Weaker correlations

Diagonal: The diagonal line represents the correlation of each variable with itself, which is always 1.

Off-Diagonal: The off-diagonal values show the correlation between different pairs of variables.

Key Observations from the Heatmap:

Strong Positive Correlations:

id and year have a very strong positive correlation, which makes sense as they are likely sequential. There are strong positive correlations between cylinders, displ, city, UCity, highway, UHighway, comb, and co2, indicating that these variables are closely related. There are also strong positive correlations between feScore and ghgScore, suggesting they are highly correlated.

Strong Negative Correlations:

cylinders, displ, city, UCity, highway, UHighway, comb, and co2 have strong negative correlations with feScore and ghgScore, indicating that vehicles with higher emissions and fuel consumption tend to have lower fuel efficiency and greenhouse gas scores.
